```python
def calculate_age_group(age):
    if 0 <= age <= 4:
        return '0-4'
    elif 5 <= age <= 9:
        return '5-9'
    elif 10 <= age <= 14:
        return '10-14'
    elif 15 <= age <= 19:
        return '15-19'
    elif 20 <= age <= 24:
        return '20-24'
    elif 25 <= age <= 29:
        return '25-29'
    elif 30 <= age <= 34:
        return '30-34'
    elif 35 <= age <= 39:
        return '35-39'
    elif 40 <= age <= 44:
        return '40-44'
    elif 45 <= age <= 49:
        return '45-49'
    elif 50 <= age <= 54:
        return '50-54'
    elif 55 <= age <= 59:
        return '55-59'
    elif 60 <= age <= 64:
        return '60-64'
    elif 65 <= age <= 69:
        return '65-69'
    elif 70 <= age <= 74:
        return '70-74'
    else:
        return '75+'

df['Age Group'] = df['Age'].apply(calculate_age_group)
```

```python
# Assuming you have a DataFrame called df with a column named "art_date" containing enrollment dates
df['art_date'] = pd.to_datetime(df['art_date'])

# Filter data for the specified date range (October 2022 to August 2023)
start_date = pd.to_datetime('2022-10-01')
end_date = pd.to_datetime('2023-08-31')
filtered_df = df[(df['art_date'] >= start_date) & (df['art_date'] <= end_date)]

# Group the data by month and count the number of enrollments in each month
monthly_enrollments = filtered_df.groupby(filtered_df['art_date'].dt.to_period("M").dt.start_time).size()

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(monthly_enrollments.index, monthly_enrollments, marker='o', linestyle='-')
plt.title('Monthly Enrollments from October 2022 to August 2023')
plt.xlabel('Month')
plt.ylabel('Number of Enrollments')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(monthly_enrollments.index, monthly_enrollments.values, color='skyblue', marker='o')
plt.title('Monthly Enrollments from October 2022 to August 2023')
plt.xlabel('Month')
plt.ylabel('Number of Enrollments')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
