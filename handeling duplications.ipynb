```python
df1 = pd.read_excel('data\\tx-curr.xlsx')
```

```python
df2 = pd.read_excel('data\\check list.xlsx')
```

```python
# Function to extract the last set of digits from alphanumeric data
def extract_numbers(data):
    matches = re.findall(r'\d+', str(data))
    if matches:
        return matches[-1]
    else:
        return ''
```

```python
df1['Extracted_Numbers'] = df1['id'].apply(extract_numbers)
```

```python
df1
```

```python
df2['Extracted_Numbers'] = df2['id'].apply(extract_numbers)
```

```python
df2
```

```python
if 'Extracted_Numbers' in df1.columns and 'Extracted_Numbers' in df2.columns:
    duplicates = df1[df1['Extracted_Numbers'].isin(df2['Extracted_Numbers'])]
```

```python
duplicates
```

```python
merged = pd.merge(df1, df2, on='Extracted_Numbers', how='left', indicator=True)
```

```python
merged
```

```python
tx_curr_only = merged[merged['_merge'] == 'left_only']
```

```python
# Define the path where you want to save the 'tx_curr_only' Excel file
output_path = 'data\\tx_curr_only.xlsx'

# Save the 'tx_curr_only' DataFrame to an Excel file
tx_curr_only.to_excel(output_path, index=False)
```
