```python
import pandas as pd
```

```python
df = pd.read_excel('data\\for sorting.xlsx')
```

```python
# Define a custom sorting function
def custom_sort(s):
    segments = s.split('/')
    last_segment = segments[-1]
    numeric_part = ''.join(filter(str.isdigit, last_segment))
    return int(numeric_part)

# Apply the custom_sort function to the 'Hospital Number' column using the .str accessor
df['numeric_sort_key'] = df['Hospital Number'].str.split('/').str[-1].apply(custom_sort)

# Sort the DataFrame based on the custom sorting key in ascending order
df = df.sort_values(by='numeric_sort_key', ignore_index=True, ascending=True)

# Drop the temporary sorting key column
df = df.drop(columns=['numeric_sort_key'])

# Display the sorted DataFrame
print(df)
```

```python
df.to_excel('sorted5.xlsx', index=False)
```
