```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Replace 'your_data.xlsx' with the actual file path of your data
df = pd.read_excel('data\\or.xlsx')
```

```python
# Filter data for people with biometrics captured
biometrics_captured = df[df['Date Biometrics Enrolled (yyyy-mm-dd)'].notna()]

# Filter data for active clients
active_clients = df[df['Current ART Status'] == 'Active']

# Count the number of people with biometrics captured and the total number of active clients
biometrics_count = len(biometrics_captured)
active_clients_count = len(active_clients)
```

```python
# Create a bar chart
plt.bar(['Biometrics Captured', 'Active Clients'], [biometrics_count, active_clients_count])
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Biometrics Captured vs. Active Clients')
plt.show()
```

```python
# Filter data for female clients
female_clients = df[(df['Sex'] == 'Female') & (df['Current ART Status'] == 'Active')]

# Filter data for female clients who have been screened for cervical cancer
screened_for_cervical_cancer = df[(df['Sex'] == 'Female') & (df['Date of Cervical Cancer Screening (yyyy-mm-dd)'].notna())]

# Count the total number of female clients and those screened for cervical cancer
total_female_clients = len(female_clients)
female_clients_screened = len(screened_for_cervical_cancer)

# Create a grouped bar chart
categories = ['Total Female Clients', 'Female Clients Screened for Cervical Cancer']
counts = [total_female_clients, female_clients_screened]

plt.bar(categories, counts)
plt.ylabel('Count')
plt.title('Total Female Clients vs. Female Clients Screened for Cervical Cancer')
plt.show()
```

```python
# Create two separate plots
plt.figure(figsize=(10, 5))

# Plot 1: Total Female Clients
plt.subplot(1, 2, 1)
plt.bar('Total Female Clients', total_female_clients, color='blue')
plt.ylabel('Count')
plt.title('Total Female Clients')

# Plot 2: Female Clients Screened for Cervical Cancer
plt.subplot(1, 2, 2)
plt.bar('Female Clients Screened for Cervical Cancer', female_clients_screened, color='green')
plt.ylabel('Count')
plt.title('Female Clients Screened for Cervical Cancer')

plt.tight_layout()
plt.show()
```

```python
 categories = ['Total Female Clients', 'Female Clients Screened for Cervical Cancer']
counts = [total_female_clients, female_clients_screened]

bar_width = 0.35
index = np.arange(len(categories))

plt.bar(index, counts, bar_width, label='active females', color=['blue', 'green'])
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Total Female Clients vs. Female Clients Screened for Cervical Cancer')
plt.xticks(index, categories)
plt.legend()
plt.tight_layout()
plt.grid('true')
plt.legend()
plt.show()
```

```python
# Create a bar chart with grouped bars
categories = ['Total Female Clients', 'Female Clients Screened for Cervical Cancer']
counts = [total_female_clients, female_clients_screened]

bar_width = 0.35
index = np.arange(len(categories))

colors = ['blue', 'green']

plt.bar(index, counts, bar_width, color=colors)
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Total Female Clients vs. Female Clients Screened for Cervical Cancer')
plt.xticks(index, categories)
# Create separate legends for active females and screened females using text annotations
plt.annotate('Active Females', xy=(index[0], counts[0]), xytext=(index[0], counts[0] - 100), color='blue', fontsize=12,
             arrowprops=dict(arrowstyle='->', color='blue'))
plt.annotate('Screened Females', xy=(index[1], counts[1]), xytext=(index[1], counts[1] - 100), color='green',
             fontsize=12, arrowprops=dict(arrowstyle='->', color='green'))

plt.tight_layout()
plt.show()



```
